```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
#plt.rcParams['figure.figsize']=(20,10)
import seaborn as sns
```

```python
data=pd.read_csv("dataset.csv")
data.head()
```

```python
x=data['Head Size(cm^3)'].values
y=data['Brain Weight(grams)'].values
sns.regplot(x = data['Head Size(cm^3)'], y = data['Brain Weight(grams)'])
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
x.shape
```

#### Now we will find m and c    
##### y=mx+c    
##### m=sum (x-mean_x)*(y-mean_y)/sum(x-mean_x)^2

```python
numer=0
denom=0
l=len(x)
x.shape
```

```python
for i in  range(l):
    numer+=(x[i]-mean_x) * (y[i]-mean_y)
    denom+=(x[i]-mean_x)**2

m=numer/denom
x.shape
```

```python
c=mean_y-(m*mean_x)
x.shape
```

```python
print(m,c)
x.shape
```

```python
ss_t=0
ss_r=0
x.shape
```

```python
for i in range(l):
      y_prid=m*(x[i])+c
      ss_t=(y_prid-mean_y)**2
      ss_r=(y[i]-mean_y)**2
r2=1-(ss_t/ss_r)
print(r2)
x.shape
```
